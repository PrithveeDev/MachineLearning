```python
# X and Y
X = df[["Vehicle Class", "Engine Size(L)", "Cylinders", "Transmission", "Fuel Type", "Fuel Consumption City (L/100 km)"]]
Y = df[["CO2 Emissions(g/km)"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder()
cols = ["Vehicle Class", "Transmission", "Fuel Type"]
X[cols] = ordinal.fit_transform(X[cols])

# maplist = [{'col': 'Fuel Type', 'mapping': {'Z': 1, 'D': 2,'N': 3, 'X': 4, 'E': 5}}]
# oe = OrdinalEncoder(mapping=maplist)
```

```python
X.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
col = ["Engine Size(L)","Cylinders", "Fuel Consumption City (L/100 km)", "Transmission"]
X[col] = scale.fit_transform(X[col])
```

```python
X.head(10)
```

```python
# train-test-split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25,random_state=0)
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
print("Training score: ",model.score(X_train,Y_train))
print("Testing score: ",model.score(X_test,Y_test))
```

```python
data = pd.DataFrame()
data["Actual values"] = Y_test.values.ravel()
data["Predicted values"] = model.predict(X_test)
data.head(10)
```
